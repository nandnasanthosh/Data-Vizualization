# copy()  and view()
# copy() -- owns the data any changes made to the copied array will not effect the original array
# view() -- doesnt owns the data any changes made to the copied array will  effect the original array

```python
# copy()
arr=np.array([1,2,3,4])
x=arr.copy()
x[2]=42
print("copied array:",x)
print("original array:",arr)
```

```python
# view
arr=np.array([1,2,3,4])
x=arr.view()
x[2]=42
print("viewed array:",x)
print("original array:",arr)
```
